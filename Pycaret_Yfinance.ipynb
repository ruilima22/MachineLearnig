##Detalhamento

```python
#ImportaÃ§Ã£o Bibliotecas
import yfinance as yf
import pandas as pd
```

```python
#Escolha da aÃ§Ã£o
df = yf.Ticker('PETR4.SA')
#Escolher intervalo de dados
petro = df.history(period = '2y')
petro
```

```python
#Retirando os campos
petro = petro.drop(['Dividends','Stock Splits'], axis=1)
petro
```

```python
#Criando novos campos
petro['MM7D'] = petro['Close'].rolling(window=7).mean().round(2)
petro['MM30D'] = petro['Close'].rolling(window=30).mean().round(2)
petro
```

```python
#5 dias para previsÃ£o
petro_prever = petro.tail(5)
petro_prever
```

```python
#Retirar os ultimos 5 dias do df
petro.drop(petro.tail(5).index, inplace=True)
petro
```

```python
#empurrar para frente o valor das aÃ§Ãµes
petro['Close'] = petro['Close'].shift(-1)
petro
```

```python
#Retirar Nulos
petro.dropna(inplace=True)
petro
```

```python
#Drop index
petro.reset_index(drop=True, inplace=True)
petro_prever.reset_index(drop=True, inplace=True)
```

```python
#import regression lib pycaret
from pycaret.regression import *
setup(data= petro, target= 'Close', session_id=123)
```

```python
#comparaÃ§Ã£o modelo
melhor_modelo = compare_models()
```

```python
print(melhor_modelo)
```

```python
#Analise do modelo
evaluate_model(melhor_modelo)
```

```python
ridge = create_model('ridge', fold=10)
```

```python
br = create_model('br', fold=10)
```

```python
#Salvando modelo
save_model(melhor_modelo, 'modelo_final')
```
